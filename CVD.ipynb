# Data Visualizations
After the data has been cleaned, the data can now be visualized. <br>
Seaborn library provides various graphs and plots for this purpose. <br>
This is necessary to understand the patterns and trends in the data. It also helps in understanding the relationships between different variables.

```python
# Set Seaborn style
sns.set(style="whitegrid")

# General Health Distribution
General_Health = data.General_Health.value_counts()
plt.figure(figsize=(8,4))
plt.title("Distribution of General Health in the Dataset")
plt.pie(General_Health, labels=General_Health.index, autopct='%.1f%%');
```

```python
# Sex Distribution in the dataset
Sex = data.Sex.value_counts()
plt.figure(figsize=(8,4))
plt.title("Male or Female")
plt.pie(Sex, labels=Sex.index, autopct='%.1f%%');
```

```python
age_order = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-64', '65-69', '70-74', '75-79', '80+']

health_order = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']

# Plot the countplot with the correct orders for both Age_Category and General_Health
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Age_Category', hue='General_Health', order=age_order, hue_order=health_order, palette='Set1')

# Title and labels
plt.title('Distribution of General Health by Age Category', fontsize=16)
plt.xlabel('Age Category', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

## Analysis for General Health:
The majority of the population reports "Good" or "Very Good" health, while a smaller proportion reports "Poor" health. This suggests the dataset is skewed towards better self-reported health.


```python
# Correlation Between Health Conditions
health_conditions = data[['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Arthritis', 'Diabetes',]]

# Convert binary categorical variables to numerical
health_conditions = health_conditions.replace({'Yes': 1, 'No': 0})

# Plot heatmap of correlations
plt.figure(figsize=(10, 6))
sns.heatmap(health_conditions.corr(), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title('Correlation Between Health Conditions', fontsize=16)
plt.tight_layout()
plt.show()
```

## Analysis for Correlation Heatmap:
The heatmap reveals significant correlations between some conditions. 
<br> These insights suggest common risk factors or co-occurrence of certain conditions.
For instance:
- Arthritis and diabetes show a moderate correlation.
- Heart disease and diabetes are also somewhat correlated, likely reflecting common comorbidities.



```python
# Bar plot showing the prevalence of Heart Disease, Diabetes, and Cancer by General Health

# Melt the relevant health conditions into long format for easier plotting
health_conditions = ['Heart_Disease', 'Diabetes', 'Skin_Cancer', 'Other_Cancer']
melted_data = data.melt(id_vars=['General_Health'], value_vars=health_conditions, var_name='Condition', value_name='Has_Condition')

# Filter the dataset for individuals who have the condition ('Yes' only)
condition_data = melted_data[melted_data['Has_Condition'] == 'Yes']

# Create a countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=condition_data, x='General_Health', hue='Condition', palette='Set1')

# Title and labels
plt.title('Prevalence of Conditions by General Health Status', fontsize=16)
plt.xlabel('General Health', fontsize=12)
plt.ylabel('Count of Individuals with Condition', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

## Analysis
This plot shows how different health conditions such as Heart Disease, Diabetes, Skin Cancer, and Other Cancer are distributed based on the general health status of individuals (Excellent, Very Good, Good, Fair, and Poor).

 - Heart Disease and Diabetes are more prevalent among individuals who report "Poor" or "Fair" general health. This makes sense since these conditions are often linked to poorer overall health.
 - Skin Cancer and Other Cancer are not as strongly linked to perceived general health as heart disease and diabetes. These conditions seem to be distributed more evenly across health statuses, although they are slightly more common in the "Fair" and "Poor" categories.
 - Interestingly, individuals who rate their health as "Excellent" or "Very Good" report significantly lower rates of all the conditions, which suggests a strong connection between self-reported health and the absence of these chronic conditions.

```python
# Scatter Plot of Weight vs Height, color-coded by Sex
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Height_(cm)', y='Weight_(kg)', hue='Sex', palette='coolwarm')

# Title and labels
plt.title('Weight vs Height, Color-Coded by Sex', fontsize=16)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

## Analysis
 - General Trend: <br>There is a positive correlation between height and weight for both males and females. Taller individuals tend to weigh more, which is an expected pattern due to larger body mass.
 - Sex-Based Differences:<br>
Men tend to be taller and heavier on average compared to women. The majority of male data points are concentrated in the higher ranges of both height and weight.
Women are generally more clustered towards the lower ranges of height and weight. However, there is some overlap between the two groups, particularly in the middle ranges.

```python
# BMI Distribution Across Heart Disease and Diabetes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Heart_Disease', y='BMI', data=data, palette="Set3")
plt.title('BMI Distribution by Heart Disease Status', fontsize=16)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Diabetes', y='BMI', data=data, palette="Set3")
plt.title('BMI Distribution by Diabetes Status', fontsize=16)
plt.tight_layout()
plt.show()
```

## Analysis for BMI and Health Conditions:
 - Individuals with heart disease tend to have a slightly higher BMI compared to those without heart disease.
 - The difference in BMI is even more pronounced for those with diabetes, indicating a strong link between obesity and diabetes.
<br><br>This suggests that weight management could be a key intervention for both conditions.


```python
# Lifestyle Factors (Alcohol, Fruit, Vegetable Consumption) vs. Health Outcomes
plt.figure(figsize=(12, 6))
sns.boxplot(x='Heart_Disease', y='Alcohol_Consumption', data=data, palette="Set2")
plt.title('Alcohol Consumption by Heart Disease Status', fontsize=16)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Heart_Disease', y='Fruit_Consumption', data=data, palette="Set2")
plt.title('Fruit Consumption by Heart Disease Status', fontsize=16)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Heart_Disease', y='Green_Vegetables_Consumption', data=data, palette="Set2")
plt.title('Green Vegetable Consumption by Heart Disease Status', fontsize=16)
plt.tight_layout()
plt.show()
```

## Analysis for Lifestyle Factors:
 - Alcohol consumption appears to be slightly lower among individuals with heart disease.
 - Fruit and vegetable consumption does not show a dramatic difference between individuals with or without heart disease.
<br><br> However, the trend is that higher vegetable consumption is slightly associated with lower heart disease rates, suggesting that diet could play a role in managing heart disease risks.
